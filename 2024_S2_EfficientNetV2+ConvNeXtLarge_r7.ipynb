#画像データに対して、10種類のラベルの1つを割り当てるモデルを作成していただきます

画像認識とは、画像および動画から文字や顔などのオブジェクトや特徴を認識し検出するパターン認識技術の一分野です。特に画像に対してラベルを割り当てる画像分類は画像認識の領域においてポピュラーな研究分野となっていて、最近は深層学習が圧倒的な精度を達成しています。今回は犬や猫などの動物や車などの無機物が写った画像に対して10種類のラベルから一つ割り当てる問題に挑戦していただきます。

本コンペを活用して、データ解析・モデル構築を体験してください。

データ概要

課題種別：分類
データ種別：画像
学習データサンプル数：5,000
画像の大きさ：96×96

```python
# ========================================
# Google Colab 上で GPU を使用する想定
# ========================================

# 1. Google Drive をマウント
from google.colab import drive
drive.mount('/content/drive')

# 2. 日本語表示のためのライブラリをインストール＆インポート
!pip install japanize_matplotlib
import japanize_matplotlib

# 3. GPUが利用可能か確認
import tensorflow as tf
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

# 4. 混合精度学習(Mixed Precision)の設定
from tensorflow.keras.mixed_precision import set_global_policy
set_global_policy('mixed_float16')
print("Mixed Precision Policy: ", tf.keras.mixed_precision.global_policy())

# 5. ファイルパスの設定
base_dir = "/content/drive/MyDrive/Colab Notebooks/AI_2024_S2"

# 6. 必要なライブラリのインポート
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.activations import swish
from tensorflow.keras.losses import CategoricalCrossentropy

!pip install scikit-learn
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
import itertools  # 混同行列の可視化に使用

# 7. データの準備（解凍＆ファイルコピー）
if not os.path.exists("train_images"):
    os.system(f"unzip -q '{base_dir}/train_images.zip'")
    os.system(f"unzip -q '{base_dir}/test_images.zip'")
    !cp "{base_dir}/label_master.tsv" .
    !cp "{base_dir}/sample_submit.tsv" .
    !cp "{base_dir}/train_master.tsv" .

# 学習用マスタ・ラベルマスタの読み込み
train_master = pd.read_csv('train_master.tsv', sep='\t')
label_master = pd.read_csv('label_master.tsv', sep='\t')
train_master['label_id'] = train_master['label_id'].astype(str)

num_classes = label_master.shape[0]  # = 10

# =========================================================
# 8. データ前処理＆学習・検証用ジェネレータの作成（共通関数）
# =========================================================
def create_generators(preprocess_func, batch_size=100, target_size=(224, 224), test_size=0.2):
    """
    画像の基本的な前処理ジェネレータ（データ拡張あり）
    ラベルが均等になるように分割する
    """
    from sklearn.model_selection import train_test_split

    train_df, val_df = train_test_split(
        train_master,
        test_size=test_size,
        random_state=42,
        stratify=train_master['label_id']
    )

    # 学習用のImageDataGenerator（データ拡張）
    datagen_train = ImageDataGenerator(
        rotation_range=45,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.2,
        zoom_range=[0.8, 1.2],
        horizontal_flip=True,
        vertical_flip=True,
        brightness_range=[0.5, 1.5],
        channel_shift_range=50.0,
        fill_mode='reflect',
        preprocessing_function=preprocess_func
    )

    # バリデーション用のImageDataGenerator（データ拡張なし）
    datagen_val = ImageDataGenerator(
        preprocessing_function=preprocess_func
    )

    train_generator = datagen_train.flow_from_dataframe(
        dataframe=train_df,
        directory="train_images",
        x_col='file_name',
        y_col='label_id',
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=True,
        seed=42
    )

    val_generator = datagen_val.flow_from_dataframe(
        dataframe=val_df,
        directory="train_images",
        x_col='file_name',
        y_col='label_id',
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=False,
        seed=42
    )

    return train_generator, val_generator

# =========================================================
# EfficientNetV2L モデル構築用
# =========================================================
from tensorflow.keras.applications.efficientnet_v2 import EfficientNetV2L, preprocess_input as preprocess_efficientnet_v2

def build_efficientnet_v2l_model(input_shape=(224, 224, 3), num_classes=10, train_base=False):
    """
    EfficientNetV2L をベースとした転移学習モデルの構築
    """
    base_model = EfficientNetV2L(
        input_shape=input_shape,
        include_top=False,
        weights='imagenet'
    )
    # ベースモデルの凍結/解凍設定
    for layer in base_model.layers:
        layer.trainable = train_base

    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dense(2048, activation=swish)(x)  # Swish活性化
    x = Dropout(0.4)(x)
    # 出力層はfloat32にキャスト
    outputs = Dense(num_classes, activation='softmax', dtype='float32')(x)

    model = Model(inputs=base_model.input, outputs=outputs)
    return model

# =========================================================
# ConvNeXtLarge モデル構築用
# =========================================================
from tensorflow.keras.applications.convnext import ConvNeXtLarge, preprocess_input as preprocess_convnext

def build_convnext_model(input_shape=(224, 224, 3), num_classes=10, train_base=False):
    """
    ConvNeXtLarge をベースとした転移学習モデルの構築
    """
    base_model = ConvNeXtLarge(
        input_shape=input_shape,
        include_top=False,
        weights='imagenet'
    )
    # ベースモデルの凍結/解凍設定
    for layer in base_model.layers:
        layer.trainable = train_base

    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dense(2048, activation=swish)(x)
    x = Dropout(0.4)(x)
    outputs = Dense(num_classes, activation='softmax', dtype='float32')(x)

    model = Model(inputs=base_model.input, outputs=outputs)
    return model

# =========================================================
# 損失関数（ラベル平滑化付き CategoricalCrossentropy）
# =========================================================
loss_fn = CategoricalCrossentropy(label_smoothing=0.1)

# --------------------------------------------------------------------
# ここから、EfficientNetV2Lモデルの学習
# --------------------------------------------------------------------
train_gen_eff, val_gen_eff = create_generators(
    preprocess_func=preprocess_efficientnet_v2,
    batch_size=100,
    target_size=(224, 224),
    test_size=0.2
)

model_eff = build_efficientnet_v2l_model(
    input_shape=(224, 224, 3),
    num_classes=num_classes,
    train_base=False
)

model_eff.compile(
    optimizer=tf.keras.optimizers.AdamW(
        learning_rate=1e-4,
        weight_decay=1e-5
    ),
    loss=loss_fn,
    metrics=['accuracy']
)

callbacks_eff = [
    EarlyStopping(monitor='val_loss', patience=9, restore_best_weights=True),
    ModelCheckpoint('best_model_efficientnetv2l_224.keras', monitor='val_loss', save_best_only=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, min_lr=1e-5)
]

print("------ EfficientNetV2L: 1st training (base frozen) ------")
history_eff = model_eff.fit(
    train_gen_eff,
    steps_per_epoch=train_gen_eff.samples // train_gen_eff.batch_size,
    validation_data=val_gen_eff,
    validation_steps=val_gen_eff.samples // val_gen_eff.batch_size,
    epochs=50,
    callbacks=callbacks_eff
)

# 可視化: 損失・精度
plt.figure()
plt.plot(history_eff.history['loss'], label='train_loss')
plt.plot(history_eff.history['val_loss'], label='val_loss')
plt.title('EfficientNetV2L: 学習(凍結)時の損失推移')
plt.legend()
plt.show()

plt.figure()
plt.plot(history_eff.history['accuracy'], label='train_accuracy')
plt.plot(history_eff.history['val_accuracy'], label='val_accuracy')
plt.title('EfficientNetV2L: 学習(凍結)時の精度推移')
plt.legend()
plt.show()

# ファインチューニング
for layer in model_eff.layers[-50:]:
    layer.trainable = True

model_eff.compile(
    optimizer=tf.keras.optimizers.AdamW(
        learning_rate=1e-4,
        weight_decay=1e-5
    ),
    loss=loss_fn,
    metrics=['accuracy']
)

fine_tune_callbacks_eff = [
    EarlyStopping(monitor='val_loss', patience=9, restore_best_weights=True),
    ModelCheckpoint('best_model_efficientnetv2l_224_finetuned.keras', monitor='val_loss', save_best_only=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, min_lr=1e-5)
]

print("------ EfficientNetV2L: 2nd training (fine-tune) ------")
history_eff_fine = model_eff.fit(
    train_gen_eff,
    steps_per_epoch=train_gen_eff.samples // train_gen_eff.batch_size,
    validation_data=val_gen_eff,
    validation_steps=val_gen_eff.samples // val_gen_eff.batch_size,
    epochs=50,
    callbacks=fine_tune_callbacks_eff
)

plt.figure()
plt.plot(history_eff_fine.history['loss'], label='train_loss')
plt.plot(history_eff_fine.history['val_loss'], label='val_loss')
plt.title('EfficientNetV2L: ファインチューニング時の損失推移')
plt.legend()
plt.show()

plt.figure()
plt.plot(history_eff_fine.history['accuracy'], label='train_accuracy')
plt.plot(history_eff_fine.history['val_accuracy'], label='val_accuracy')
plt.title('EfficientNetV2L: ファインチューニング時の精度推移')
plt.legend()
plt.show()

# ベストモデルをロード
best_model_eff = load_model(
    'best_model_efficientnetv2l_224_finetuned.keras',
    custom_objects={'swish': swish}
)

# --------------------------------------------------------------------
# ここから、ConvNeXtLargeモデルの学習
# --------------------------------------------------------------------
train_gen_conv, val_gen_conv = create_generators(
    preprocess_func=preprocess_convnext,
    batch_size=100,
    target_size=(224, 224),
    test_size=0.2
)

model_conv = build_convnext_model(
    input_shape=(224, 224, 3),
    num_classes=num_classes,
    train_base=False
)

model_conv.compile(
    optimizer=tf.keras.optimizers.AdamW(
        learning_rate=1e-4,
        weight_decay=1e-5
    ),
    loss=loss_fn,
    metrics=['accuracy']
)

callbacks_conv = [
    EarlyStopping(monitor='val_loss', patience=9, restore_best_weights=True),
    ModelCheckpoint('best_model_convnextlarge_224.keras', monitor='val_loss', save_best_only=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, min_lr=1e-5)
]

print("------ ConvNeXtLarge: 1st training (base frozen) ------")
history_conv = model_conv.fit(
    train_gen_conv,
    steps_per_epoch=train_gen_conv.samples // train_gen_conv.batch_size,
    validation_data=val_gen_conv,
    validation_steps=val_gen_conv.samples // val_gen_conv.batch_size,
    epochs=50,
    callbacks=callbacks_conv
)

# 可視化: 損失・精度
plt.figure()
plt.plot(history_conv.history['loss'], label='train_loss')
plt.plot(history_conv.history['val_loss'], label='val_loss')
plt.title('ConvNeXtLarge: 学習(凍結)時の損失推移')
plt.legend()
plt.show()

plt.figure()
plt.plot(history_conv.history['accuracy'], label='train_accuracy')
plt.plot(history_conv.history['val_accuracy'], label='val_accuracy')
plt.title('ConvNeXtLarge: 学習(凍結)時の精度推移')
plt.legend()
plt.show()

# ファインチューニング
for layer in model_conv.layers[-50:]:
    layer.trainable = True

model_conv.compile(
    optimizer=tf.keras.optimizers.AdamW(
        learning_rate=1e-4,
        weight_decay=1e-5
    ),
    loss=loss_fn,
    metrics=['accuracy']
)

fine_tune_callbacks_conv = [
    EarlyStopping(monitor='val_loss', patience=9, restore_best_weights=True),
    ModelCheckpoint('best_model_convnextlarge_224_finetuned.keras', monitor='val_loss', save_best_only=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, min_lr=1e-5)
]

print("------ ConvNeXtLarge: 2nd training (fine-tune) ------")
history_conv_fine = model_conv.fit(
    train_gen_conv,
    steps_per_epoch=train_gen_conv.samples // train_gen_conv.batch_size,
    validation_data=val_gen_conv,
    validation_steps=val_gen_conv.samples // val_gen_conv.batch_size,
    epochs=50,
    callbacks=fine_tune_callbacks_conv
)

plt.figure()
plt.plot(history_conv_fine.history['loss'], label='train_loss')
plt.plot(history_conv_fine.history['val_loss'], label='val_loss')
plt.title('ConvNeXtLarge: ファインチューニング時の損失推移')
plt.legend()
plt.show()

plt.figure()
plt.plot(history_conv_fine.history['accuracy'], label='train_accuracy')
plt.plot(history_conv_fine.history['val_accuracy'], label='val_accuracy')
plt.title('ConvNeXtLarge: ファインチューニング時の精度推移')
plt.legend()
plt.show()

# ベストモデルをロード
best_model_conv = load_model(
    'best_model_convnextlarge_224_finetuned.keras',
    custom_objects={'swish': swish}
)

# --------------------------------------------------------------------
# 混同行列の可視化関数（まとめて利用）
# --------------------------------------------------------------------
def plot_confusion_matrix(cm, title, classes=None):
    plt.figure(figsize=(6, 5))
    plt.imshow(cm, interpolation='nearest', cmap='Blues')
    plt.title(title)
    plt.colorbar()
    if classes is None:
        classes = [str(i) for i in range(num_classes)]
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 color="black" if cm[i, j] > thresh else "black")
    plt.ylabel('正解ラベル')
    plt.xlabel('予測ラベル')
    plt.tight_layout()
    plt.show()

# --------------------------------------------------------------------
# それぞれのモデル単独での評価（バリデーション）
# --------------------------------------------------------------------
val_gen_eff.reset()  # バリデーションジェネレータを最初から
val_predictions_eff = best_model_eff.predict(val_gen_eff, steps=len(val_gen_eff), verbose=1)
val_pred_id_eff = np.argmax(val_predictions_eff, axis=1)
true_classes_eff = val_gen_eff.classes

acc_eff = np.mean(true_classes_eff == val_pred_id_eff)
print(f"EfficientNet 単独モデル Validation Accuracy: {acc_eff:.4f}")

cm_eff = confusion_matrix(true_classes_eff, val_pred_id_eff)
plot_confusion_matrix(cm_eff, "EfficientNet 単独モデルの混同行列")

val_gen_conv.reset()
val_predictions_conv = best_model_conv.predict(val_gen_conv, steps=len(val_gen_conv), verbose=1)
val_pred_id_conv = np.argmax(val_predictions_conv, axis=1)
true_classes_conv = val_gen_conv.classes

acc_conv = np.mean(true_classes_conv == val_pred_id_conv)
print(f"ConvNeXt 単独モデル Validation Accuracy: {acc_conv:.4f}")

cm_conv = confusion_matrix(true_classes_conv, val_pred_id_conv)
plot_confusion_matrix(cm_conv, "ConvNeXt 単独モデルの混同行列")

# --------------------------------------------------------------------
# ★追加・変更★ 単独モデル評価結果をファイル出力
# バリデーションにおける「ファイル名」「正解ラベルID」「予測ラベルID」「(正誤フラグ)」などを保存
# --------------------------------------------------------------------
val_filenames_eff = val_gen_eff.filenames
df_eff_result = pd.DataFrame({
    "file_name": val_filenames_eff,
    "true_label": true_classes_eff,
    "pred_label": val_pred_id_eff
})
df_eff_result["correct"] = (df_eff_result["true_label"] == df_eff_result["pred_label"]).astype(int)

output_eff_val = f"{base_dir}/evaluation_efficientnet_0406_r0.tsv"
df_eff_result.to_csv(output_eff_val, sep="\t", index=False)
print(f"EfficientNet 単独モデルの評価結果を {output_eff_val} に出力しました。")

val_filenames_conv = val_gen_conv.filenames
df_conv_result = pd.DataFrame({
    "file_name": val_filenames_conv,
    "true_label": true_classes_conv,
    "pred_label": val_pred_id_conv
})
df_conv_result["correct"] = (df_conv_result["true_label"] == df_conv_result["pred_label"]).astype(int)

output_conv_val = f"{base_dir}/evaluation_convnext_0406_r0.tsv"
df_conv_result.to_csv(output_conv_val, sep="\t", index=False)
print(f"ConvNeXt 単独モデルの評価結果を {output_conv_val} に出力しました。")


# --------------------------------------------------------------------
# バリデーションデータでの予測（各モデル → アンサンブル）
# --------------------------------------------------------------------
val_gen_eff.reset()
val_predictions_eff = best_model_eff.predict(val_gen_eff, steps=len(val_gen_eff), verbose=1)
val_gen_conv.reset()
val_predictions_conv = best_model_conv.predict(val_gen_conv, steps=len(val_gen_conv), verbose=1)

val_pred_id_eff = np.argmax(val_predictions_eff, axis=1)
val_pred_id_conv = np.argmax(val_predictions_conv, axis=1)

# 同じく shuffle=False の場合、ジェネレータの並びが揃っている想定
true_classes_eff = val_gen_eff.classes
true_classes_conv = val_gen_conv.classes

# --------------------------------------------------------------------
# アンサンブル評価を複数手法で実施
# --------------------------------------------------------------------
# 1) 確率平均
val_prob_ensemble_avg = 0.5 * val_predictions_eff + 0.5 * val_predictions_conv
val_pred_id_ensemble_avg = np.argmax(val_prob_ensemble_avg, axis=1)

# 2) 重み付き確率平均（例：EfficientNet 0.4, ConvNeXt 0.6）
val_prob_ensemble_w = 0.4 * val_predictions_eff + 0.6 * val_predictions_conv
val_pred_id_ensemble_w = np.argmax(val_prob_ensemble_w, axis=1)

# 3) ハード投票
val_pred_id_ensemble_hard = []
for i in range(len(val_predictions_eff)):
    eff_c = val_pred_id_eff[i]
    conv_c = val_pred_id_conv[i]
    if eff_c == conv_c:
        val_pred_id_ensemble_hard.append(eff_c)
    else:
        max_prob_eff = np.max(val_predictions_eff[i])
        max_prob_conv = np.max(val_predictions_conv[i])
        if max_prob_eff >= max_prob_conv:
            val_pred_id_ensemble_hard.append(eff_c)
        else:
            val_pred_id_ensemble_hard.append(conv_c)
val_pred_id_ensemble_hard = np.array(val_pred_id_ensemble_hard)

# --------------------------------------------------------------------
# バリデーション用の混同行列を表示し、ヒートマップ作成（各アンサンブル手法）
# --------------------------------------------------------------------
true_classes = true_classes_eff  # EfficientNetのval_genを基準とする

cm_avg = confusion_matrix(true_classes, val_pred_id_ensemble_avg)
plot_confusion_matrix(cm_avg, "アンサンブル(確率平均)の混同行列")

cm_w = confusion_matrix(true_classes, val_pred_id_ensemble_w)
plot_confusion_matrix(cm_w, "アンサンブル(重み付き確率平均)の混同行列")

cm_hard = confusion_matrix(true_classes, val_pred_id_ensemble_hard)
plot_confusion_matrix(cm_hard, "アンサンブル(ハード投票)の混同行列")

acc_avg = np.mean(true_classes == val_pred_id_ensemble_avg)
acc_w = np.mean(true_classes == val_pred_id_ensemble_w)
acc_hard = np.mean(true_classes == val_pred_id_ensemble_hard)

print(f"Ensemble(確率平均) Validation Accuracy: {acc_avg:.4f}")
print(f"Ensemble(重み付き確率平均) Validation Accuracy: {acc_w:.4f}")
print(f"Ensemble(ハード投票) Validation Accuracy: {acc_hard:.4f}")

# --------------------------------------------------------------------
# テストデータでの推論（両モデル → アンサンブル）
# --------------------------------------------------------------------
test_filenames = os.listdir("test_images")
test_df = pd.DataFrame({'file_name': test_filenames})

test_datagen_eff = ImageDataGenerator(
    preprocessing_function=preprocess_efficientnet_v2
)
test_generator_eff = test_datagen_eff.flow_from_dataframe(
    dataframe=test_df,
    directory="test_images",
    x_col='file_name',
    y_col=None,
    target_size=(224,224),
    batch_size=100,
    class_mode=None,
    shuffle=False
)

test_datagen_conv = ImageDataGenerator(
    preprocessing_function=preprocess_convnext
)
test_generator_conv = test_datagen_conv.flow_from_dataframe(
    dataframe=test_df,
    directory="test_images",
    x_col='file_name',
    y_col=None,
    target_size=(224,224),
    batch_size=100,
    class_mode=None,
    shuffle=False
)

pred_eff = best_model_eff.predict(test_generator_eff, steps=len(test_generator_eff), verbose=1)
pred_conv = best_model_conv.predict(test_generator_conv, steps=len(test_generator_conv), verbose=1)

# 1) 確率平均
pred_avg = 0.5 * pred_eff + 0.5 * pred_conv
test_pred_id_avg = np.argmax(pred_avg, axis=1)

# 2) 重み付き確率平均 (0.4 vs 0.6)
pred_w = 0.4 * pred_eff + 0.6 * pred_conv
test_pred_id_w = np.argmax(pred_w, axis=1)

# 3) ハード投票
test_pred_id_hard = []
for i in range(len(pred_eff)):
    eff_c = np.argmax(pred_eff[i])
    conv_c = np.argmax(pred_conv[i])
    if eff_c == conv_c:
        test_pred_id_hard.append(eff_c)
    else:
        max_prob_eff = np.max(pred_eff[i])
        max_prob_conv = np.max(pred_conv[i])
        if max_prob_eff >= max_prob_conv:
            test_pred_id_hard.append(eff_c)
        else:
            test_pred_id_hard.append(conv_c)
test_pred_id_hard = np.array(test_pred_id_hard)

# --------------------------------------------------------------------
# ファイル出力（アンサンブル結果を複数ファイルに）
# ★変更★: 0405 -> 0406 へ変更
# --------------------------------------------------------------------
def create_submission(file_names, pred_label_ids, output_path):
    submission = pd.DataFrame({
        'file_name': file_names,
        'label_id': pred_label_ids
    })
    # ソート用にファイル名の数字部分を抽出
    submission['file_num'] = submission['file_name'].str.extract(r'(\d+)').astype(int)
    submission = submission.sort_values(by='file_num').drop(columns='file_num')
    submission.to_csv(output_path, sep='\t', index=False, header=False)
    print(f"提出用ファイル {output_path} を作成しました。")

output_avg = f'{base_dir}/submission_EFV2CNXL_ensemble_avg_224_0406_r0.tsv'
create_submission(test_filenames, test_pred_id_avg, output_avg)

output_w = f'{base_dir}/submission_EFV2CNXL_ensemble_weighted_224_0406_r0.tsv'
create_submission(test_filenames, test_pred_id_w, output_w)

output_hard = f'{base_dir}/submission_EFV2CNXL_ensemble_hardvote_224_0406_r0.tsv'
create_submission(test_filenames, test_pred_id_hard, output_hard)

print("アンサンブル評価が完了しました。")
```
